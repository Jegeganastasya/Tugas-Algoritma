```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
#Akan dibuat 2 persamaan garis dan solusi yang mungkin untuk persamaan garis tersebut
#Persamaan garis pertama melewati dua titik yaitu (a,b) dan (c,d)
a_38 = input("x1")
b_38 = input("y1")
c_38 = input("x2")
d_38 = input("y2")
m1_38 = (float(d_38)-float(b_38))/(float(c_38)-float(a_38))
print ("Gradien pertama adalah {:.2f}".format(m1_38))

if float(-m1_38)*float(a_38)+float(b_38)>0:
    print("Persamaan garis pertama adalah y1={:.2f}x+{:.2f}".format(m1_38,float(-m1_38)*float(a_38)+float(b_38)))
else:
    print("Persamaan garis pertama adalah y1={:.2f}x{:.2f}".format(m1_38,m1_38*a_38+b_38))
#y1=m1_38*(x-a_38)+b_38  
    
 #Persamaan garis kedua melewati dua titik yaitu (e,f) dan (g,h)  
e_38 = input("x1")
f_38 = input("y1")
g_38 = input("x2")
h_38 = input("y2")
m2_38 = (float(h_38)-float(f_38))/(float(g_38)-float(e_38))
print ("Gradien kedua adalah {:.2f}".format(m2_38))

if float(-m2_38)*float(e_38)+float(f_38)>0:
    print("Persamaan garis pertama adalah y2={:.2f}x+{:.2f}".format(m2_38,float(-m2_38)*float(e_38)+float(f_38)))
else:
    print("Persamaan garis pertama adalah y2={:.2f}x{:.2f}".format(m2_38,m2_38*e_38+f_38))
#y2=m2_38*(x-e_38)+f_38  


if m1_38==m2_38:
    print("Gradiennya sama, mungkin solusinya tidak terbatas")
else:
    
    # Titik potong kedua garis pada (tp_x,tp_y)
    tp_x_38 = (float(m1_38)*float(a_38)-float(b_38)-float(m2_38)*float(e_38)+float(f_38))/(float(m1_38)-float(m2_38))
    tp_y_38 = float(m2_38)*(float(tp_x_38)-float(e_38))+float(f_38)

    print("Titik potong kedua garis adalah ({:.2f},{:.2f})".format(tp_x_38,tp_y_38))

    x_38=np.arange(float(tp_x_38)-10,float(tp_x_38)+10)
    plt.figure(figsize=(5,5),dpi=600)
    if float(-m1_38)*float(a_38)+float(b_38)>0:
        plt.plot(float(x_38),float(m1_38)*(float(x_38)-float(a_38))+float(b_38),color='red'label=',y={:.2f}x+{:.2f}'.format(m1_38,float(-m1_38)*float(a_38)+float(b_38))
                 print.plot()
    else:
        plt.plot(float(x_38),float(m1_38)*(float(x_38)-float(a_38))+float(b_38),color='red',label='y={:.2f}x{:.2f}'.format(m1_38,float(-m1_38)*float(a_38)+float(b_38))
    if float(-m2_38)*float(e_38)+float(f_38)>0:
        plt.plot(x_38,float(m2_38)*(float(x_38)-float(e_38))+float(f_38),color='blue',label='y={:.2f}x+{:.2f}'.format(m2_38,float(-m2_38)*float(e_38)+float(f_38))
    else:
        plt.plot(x_38,float(m2_38)*(float(x_38)-float(e_38))+float(f_38),color='blue',label='y={:.2f}x{:.2f}'.format(m2_38,float(-m2_38)*float(e_38)+float(f_38))
        
    plt.plot(tp_x_38,tp_y_38,color='green',marker='o',markersize=5)
    plt.legend(loc='best')
    plt.title("Persamaan Garis")
    plt.show()
    

```
