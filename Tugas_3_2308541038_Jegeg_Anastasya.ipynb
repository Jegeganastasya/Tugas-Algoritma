```python
berat = float(input("Masukan berat badan anda!"))
kilo_gram = input("Dalam kilogram atau gram?")
if kilo_gram.upper() == 'GRAM':
    berat2 = berat / 1000
    print ("Berat badanmu dalam kilogram adalah:",berat2, "kg") 
elif kilo_gram.upper() == 'KILOGRAM':
    print (berat, "kg" )

tinggi = float(input("Masukan tinggi badanmu"))
tinggi2 = input ("Dalam inci atau cm?")
if tinggi2.upper() == 'INCI':
    tinggi3 = tinggi * 0,393701
    print ("Tinggi badanmu dalam cm adalah",tinggi3,"cm")
elif tinggi == ('CM'):
    print (tinggi,"cm")
tinggi_cm = float(input("Masukkan tinggi dalam cm"))
print (tinggi_cm, "cm")

laki_wanita = ("Masukkan jenis kelamin kamu")
laki_wanita2 = input ("Perempuan atau Laki-laki?")
if laki_wanita2.upper() == 'PEREMPUAN':
    badan_ideal_wanita = (float(tinggi_cm) - 100) - float(tinggi_cm) - 100 * 0,15
    print(badan_ideal_wanita)
elif laki_wanita2 == 'LAKI-LAKI':
    badan_ideal_laki = float(tinggi_cm - 100) - float((tinggi_cm - 100) * 0,10)
    print(badan_ideal_laki)
```
