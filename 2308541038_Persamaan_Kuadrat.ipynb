```python
import numpy as np
import matplotlib.pyplot as plt
import cmath # menangani akar akar kuadrat yang negatif

# Menentukan persamaan kuadrat jika diketahui akar akarnya
# Menentukan grafik fungsi kuadrat
# Menentukan titik puncak dan sumbu simetri
# Menentukan posisi titik puncak (di kiri sumbu y, pada sumbu y (b = 0), di kanan sumbu y)
# Bagaimana kondisi parabola memotong sumbu y, (potong sumbu y positif, potong sumbu y di pusat(0,0), dan potong sumbu y ngatif)
# Macam macam akar persamaan kuadrat dapat menggunakan rumus ABC
# Menyusun persamaan kuadrat baris jika diketahui akar akarnya

print("Belajar persamaan kuadrat")
print()
#input atau masukan koefisien a, b dan konstanta c sehingga memenuhi persamaan ax^2 + bx + c = 0
print("Persamaan yang dibuat akan memenuhi ax^2 + bx + c = 0")
a_38 = float(input("Masukan koefisien nilai a persamaan pertama"))

if a_38==0:
    print("Koefisien a harus bukan sama dengan nol")

else:
    b_38 = float(input("Masukan koefisien nilai b persamaan pertama"))
    c_38 = float(input("Masukkan konstanta nilai c persamaan pertama"))
    print("Maka persamaan nya akan berbentuk {:.2f}x^2+{:.2f}x+{:.2f}=0".format(a_38,b_38,c_38))
    if a_38>0:
        print("Karena koefisien a positif maka kurva akan terbuka keatas")
    elif a_38<0:
        print("Karena koefisien a ngatif maka kurva akan terbuka kebawah")
    
    #setelah diketahui persamaannya maka akan dicari nilai diskriminan 
    
    D_38 = b_38*b_38-4*a_38*c_38
    print ("Diskriminan nya adalah D = {:.2f}".format(D_38))
    
    #setelah diketahui nilai diskriminan nya akan dicari bentuk bentuk akarnya
    # ada akar kembar jika D = 0, D<0 menandakan imajiner dan ada dua kemungkinan yaitu + dan -, D>0 menandakan bilangan reel dan ada dua kemungkinannya yaitu + dan -
    if D_38==0:
        akar1_38 = (-b_38-cmath.sqrt(D_38))/(2*a_38)
        print("Akarnya kembar yaitu x1=x2={:.2f}".format(akar1_38))
        x = np.linspace(akar1_38-10,akar1_38+10,100)
        y_38 = a_38*x**2 + b_38*x + c_38
        fig = plt.figure()
        ax = fig.add_subplot(1, 1, 1)

        # gambar fungsinya
        plt.plot(x,y_38, 'r')

        # show grafiknya
        plt.show()


        koordtpx_38 = -b_38/2*a_38
        koordtpy_38 = a_38*koordtpx_38*koordtpx_38+b_38*koordtpx_38+c_38
        print("Jadi koordinat titik puncaknya adalah ({:.2f},{:.2f})".format(koordtpx_38,koordtpy_38))
        tpthdpy_38 = a_38*0+b_38*0+c_38
        print("Maka titik potong terhadap sumbu y adalah (0,{:.2f})".format(tpthdpy_38))
    else:
        if D_38<0:
            print("Diskriminan nya imajiner sehingga ada dua akar yaitu x1 dan x2")
            akar1_38 = (-b_38-cmath.sqrt(D_38))/(2*a_38) 
            akar2_38 = (-b_38+cmath.sqrt(D_38))/(2*a_38)
            print("x1 = {:.2f} dan x2 = {:.2f}".format(akar1_38,akar2_38))
            x = np.linspace(akar1_38-10,akar1_38+10,100)
            y_38 = a_38*x**2 + b_38*x + c_38
            fig = plt.figure()
            ax = fig.add_subplot(1, 1, 1)

            # gambar fungsinya
            plt.plot(x,y_38, 'r')

            # show grafiknya
            plt.show()

            koordtpx_38 = -b_38/2*a_38
            koordtpy_38 = a_38*koordtpx_38*koordtpx_38+b_38*koordtpx_38+c_38
            print("Jadi koordinat titik puncaknya adalah ({:.2f},{:.2f})".format(koordtpx_38,koordtpy_38))
            tpthdpy_38 = a_38*0+b_38*0+c_38
            print("Maka titik potong terhadap sumbu y adalah (0,{:.2f})".format(tpthdpy_38))
           
        else:
            print("Diskriminan nya real berbeda sehingga ada dua akar yaitu x1 dan x2")
            akar1_38 = (-b_38-cmath.sqrt(D_38))/(2*a_38) 
            akar2_38 = (-b_38+cmath.sqrt(D_38))/(2*a_38)
            print("x1 = {:.2f} dan x2 = {:.2f}".format(akar1_38,akar2_38))
            x = np.linspace(akar1_38-10,akar1_38+10,100)
            y = a_38*x**2 + b_38*x + c_38
            fig = plt.figure()
            ax = fig.add_subplot(1, 1, 1)

            # gambar fungsinya
            plt.plot(x,y_38, 'r')

            # show grafiknya
            plt.show()
            koordtpx_38 = -b_38/2*a_38
            koordtpy_38 = a_38*koordtpx_38*koordtpx_38+b_38*koordtpx_38+c_38
            print("Jadi koordinat titik puncaknya adalah ({:.2f},{:.2f})".format(koordtpx_38,koordtpy_38))
            tpthdpy_38 = a_38*0+b_38*0+c_38
            print("Maka titik potong terhadap sumbu y adalah (0,{:.2f})".format(tpthdpy_38))

    # jika outputnya mengandung j maka artinya itu bilangan riil
    # setelah mengetahui akar akarnya maka akan dibuat grafik persamaan kuadrat nya
    
```
